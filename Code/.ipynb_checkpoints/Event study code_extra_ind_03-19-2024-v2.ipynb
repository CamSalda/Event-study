```python
import pandas as pd
import numpy as np
from pandas_datareader import data as pdr
import datetime as dt
import statsmodels.api as sm
import yfinance as yfin
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import chardet
import math
```

## 1. Importing data<a id="chapter1"></a>

First I define the dates for the analysis

```python
start = dt.date(2017,12,31)
end = dt.date(2022,12,31)
```

I import the files for the indexes and the Risk Market Premium

```python
url = "https://github.com/CamSalda/Inflation-Reduction-Act-event-study/blob/main/F-F_Research_Data_Factors_daily.csv?raw=true"

RF = pd.read_csv(url)[["Date", "Mkt-RF", "RF"]]
RF["Mkt"] = RF["Mkt-RF"] + RF["RF"]
RF["Mkt"]= RF["Mkt"]/100
RF 
```

```python
RF["Date"] = pd.to_datetime(RF["Date"], format ="%Y%m%d").dt.strftime("%Y-%m-%d")

RF
```

Now, I import the "^SP500_auto" Ticker.

```python
tickers = ["^SP500-251020"]
```

```python
yfin.pdr_override()

elec_v_prices = pdr.get_data_yahoo(tickers, start, end)["Adj Close"]
elec_v_prices = pd.DataFrame(elec_v_prices)
elec_v_prices.reset_index(inplace = True)
elec_v_prices
```

```python
elec_v_prices.rename(columns={"Adj Close": "^SP500_auto"}, inplace = True )
```

Importing the Nasdaq Clean Energy Edge Index

```python
cegei = pd.read_csv("https://github.com/CamSalda/Inflation-Reduction-Act-event-study/blob/main/NASDAQ%20Clean%20Edge%20Green%20Energy%20Index.csv?raw=true", 
                    index_col="Date")["Close/Last"]
```

```python
cegei.index = pd.to_datetime(cegei.index)
cegei = cegei.sort_index()
cegei = cegei.loc[start:end]
cegei =pd.DataFrame(cegei).rename(columns = {"Close/Last": "NASDAQ CEGEI"})
```

Importing the Dow Jones U.S. Oil & Gas Index

```python
with open("Dow Jones U.S. Oil & Gas Index.csv", 'rb') as f:
    enc = chardet.detect(f.read())  # or readline if the file is large
    
dj_ogi = pd.read_csv("https://github.com/CamSalda/Inflation-Reduction-Act-event-study/blob/main/Dow%20Jones%20U.S.%20Oil%20%26%20Gas%20Index.csv?raw=true", 
                     encoding = enc['encoding'])
```

```python
dj_ogi.rename(columns = {"Effective date ": "Date", "Dow Jones U.S. Oil & Gas Index": "DJ usogi"}, inplace= True)
dj_ogi["Date"] = pd.to_datetime(dj_ogi["Date"])
dj_ogi.set_index("Date", inplace = True)
dj_ogi = dj_ogi.loc[start:end]
```

Importing the S&P500 Energy index

```python
SP500_ei =  pd.read_csv("https://github.com/CamSalda/Inflation-Reduction-Act-event-study/blob/main/S%26P500%20Energy%20Index.csv?raw=true")
SP500_ei.rename(columns = {"Effective date ": "Date", "S&P 500 Energy (Sector)": "SP500_ei"}, inplace= True)
SP500_ei["Date"] = pd.to_datetime(SP500_ei["Date"])
SP500_ei.set_index("Date", inplace = True)
SP500_ei = SP500_ei.loc[start:end]
```

Importing the S&P Kensho Electric Vehicles Index

```python
k_ev = pd.read_csv("https://github.com/CamSalda/Inflation-Reduction-Act-event-study/blob/main/S%26P%20Kensho%20Electric%20Vehicles%20Index.csv?raw=true")
k_ev.rename(columns = {"Effective date ": "Date", "S&P Kensho Electric Vehicles Index (USD)": "Kensho_evi"}, inplace= True)
k_ev["Date"] = pd.to_datetime(k_ev["Date"])
k_ev.set_index("Date", inplace = True)
k_ev = k_ev.loc[start:end]
```

# 2. Data preparation<a id="chapter2"></a>

## 2.1 Merging data <a id="section_2_1"></a>

```python
RF["Date"] = pd.to_datetime(RF["Date"])
elec_v_prices["Date"] = pd.to_datetime(elec_v_prices["Date"])
```

```python
DF1 =pd.merge(elec_v_prices, RF , how ="inner", on = "Date")
```

```python
DF2 = pd.merge(DF1, cegei , how ="inner", on = "Date")
DF2 = pd.merge(DF2, dj_ogi, how ="inner", on = "Date")
DF2 = pd.merge(DF2, SP500_ei, how ="inner", on = "Date")
DF2 = pd.merge(DF2, k_ev, how ="inner", on = "Date")
DF2
```

## 2.2 Creating the dummy variables <a id="section_2_2"></a>

First, I create the dummies for the analysis of the impact after the event, with a time window of two days:


```python
DF2["sept_24_to_sept_28_2021"] = 0
DF2["dec_17_to_dec_21_2021"] = 0
DF2["jul_26_to_jul_28_2022"] = 0
DF2["aug_05_to_aug_09_2022"] = 0

mask1 = ((DF2["Date"]=="2021-09-24") | (DF2["Date"]=="2021-09-27") | (DF2["Date"]=="2021-09-28")  )
mask2 = ((DF2["Date"]=="2021-12-17") | (DF2["Date"]=="2021-12-20") | (DF2["Date"]=="2021-12-21")  )
mask3 = ((DF2["Date"]=="2022-07-26") | (DF2["Date"]=="2022-07-27") | (DF2["Date"]=="2022-07-28")  )
mask4 = ((DF2["Date"]=="2022-08-05") | (DF2["Date"]=="2022-08-08") | (DF2["Date"]=="2022-08-09")  )

DF2.loc[mask1, "sept_24_to_sept_28_2021"] = 1
DF2.loc[mask2, "dec_17_to_dec_21_2021"] = 1
DF2.loc[mask3, "jul_26_to_jul_28_2022"] = 1
DF2.loc[mask4, "aug_05_to_aug_09_2022"] = 1


print(DF2["sept_24_to_sept_28_2021"].value_counts())
print(DF2["dec_17_to_dec_21_2021"].value_counts())
print(DF2["jul_26_to_jul_28_2022"].value_counts())
print(DF2["aug_05_to_aug_09_2022"].value_counts())
```

Then, I create the dummies for the analysis of the impact after the event, with a time window of five days:


```python
DF2["sept_23_to_sept_29_2021"] = 0
DF2["dec_16_to_dec_22_2021"] = 0
DF2["jul_25_to_jul_29_2029"] = 0
DF2["aug_04_to_aug_10_2022"] = 0

mask1 = ((DF2["Date"]=="2021-09-23") | (DF2["Date"]=="2021-09-24") | (DF2["Date"]=="2021-09-27") | 
         (DF2["Date"]=="2021-09-28") | (DF2["Date"]=="2021-09-29"))

mask2 = ((DF2["Date"]=="2021-12-16") | (DF2["Date"]=="2021-12-17") | (DF2["Date"]=="2021-12-20") | 
         (DF2["Date"]=="2021-12-21") | (DF2["Date"]=="2021-12-22") )

mask3 = ((DF2["Date"]=="2022-07-25") | (DF2["Date"]=="2022-07-26") | (DF2["Date"]=="2022-07-27") | 
         (DF2["Date"]=="2022-07-28") | (DF2["Date"]=="2022-07-29") )

mask4 = ((DF2["Date"]=="2022-08-04") | (DF2["Date"]=="2022-08-05") | (DF2["Date"]=="2022-08-08") | 
         (DF2["Date"]=="2022-08-09") | (DF2["Date"]=="2022-08-10") )

DF2.loc[mask1, "sept_23_to_sept_29_2021"] = 1
DF2.loc[mask2, "dec_16_to_dec_22_2021"] = 1
DF2.loc[mask3, "jul_25_to_jul_29_2029"] = 1
DF2.loc[mask4, "aug_04_to_aug_10_2022"] = 1


print(DF2["sept_23_to_sept_29_2021"].value_counts())
print(DF2["dec_16_to_dec_22_2021"].value_counts())
print(DF2["jul_25_to_jul_29_2029"].value_counts())
print(DF2["aug_04_to_aug_10_2022"].value_counts())
```

## 2.3 Calculating returns <a id="section_2_3"></a>

To obtain the returns of the ECO Index and the S&P 500 Automobiles Industry index, I apply the log transformation and then obtain the first differences.

```python
#Apply log transformation
DF2["ln^SP500_auto"] = np.log(DF2["^SP500_auto"])
DF2["ln_NASDAQ CEGEI"] = np.log(DF2["NASDAQ CEGEI"])
DF2["ln_DJ usogi"] = np.log(DF2["DJ usogi"])
DF2["ln_SP500_ei"] = np.log(DF2["SP500_ei"])
DF2["ln_Kensho_evi"] = np.log(DF2["Kensho_evi"])


#Apply first diff
DF2["Ret^SP500_auto"] = DF2["ln^SP500_auto"].diff() 
DF2["Ret_NASDAQ CEGEI"] = DF2["ln_NASDAQ CEGEI"].diff() 
DF2["Ret_Kensho_evi"] = DF2["ln_Kensho_evi"].diff() 
DF2["Ret_DJ usogi"] = DF2["ln_DJ usogi"].diff() 
DF2["Ret_SP500_ei"] = DF2["ln_SP500_ei"].diff() 


#Drop first observation
DF2.drop(0, inplace = True)
```

Now, I obtain the descriptive statistics

```python
vars= ["Mkt", 'Ret_NASDAQ CEGEI',"Ret^SP500_auto", 'Ret_Kensho_evi','Ret_DJ usogi', 'Ret_SP500_ei']

sum_stats = DF2[vars].describe()
```

```python
#Obtain kurtosis
kurtosis = pd.Series()

for x in vars:
    kurtosis[str(x)] = DF2[str(x)].kurt()
    
    
kurtosis = kurtosis.to_frame().rename(columns= {0:"kurtosis"}).transpose()



#Obtain skewness
skewness = pd.Series()

for x in vars:
    skewness[str(x)] = DF2[str(x)].skew()
    
    
skewness = skewness.to_frame().rename(columns= {0:"skewness"}).transpose()
```

```python
sum_stats = pd.concat([sum_stats, kurtosis, skewness])

sum_stats.to_csv('summary_stats.csv')

sum_stats
```

Now I apply the Jarque Bera test

```python
norm = pd.DataFrame()

for x in sum_stats.columns:
    norm[x] = stats.jarque_bera(DF2[x].dropna())

norm.index = ["Statistic", "p-value"]


norm.to_csv('Jarque_Bera.csv')
norm
```

```python
fig, axs = plt.subplots(3,2, figsize=(20, 18))


for x in sum_stats.columns:
    a = math.floor(sum_stats.columns.get_loc(x)/2)
    b = sum_stats.columns.get_loc(x)%2
    
    sns.histplot(DF2[x], kde = True, color = "#3498db", ax=axs[a,b])
    axs[a,b].set_title(x, fontsize = 14)
    axs[a,b].set_xlabel('')
    

plt.savefig('graph4.png')

fig.tight_layout()
```
